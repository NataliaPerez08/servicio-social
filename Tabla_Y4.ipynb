```python
import os
from specdal import Collection, Spectrum, read
from matplotlib import pyplot as plt
import pandas as pd

datadir = "/content/drive/MyDrive/Colab Notebooks/Espectros_FORS_2/Y4"
for f in sorted(os.listdir(datadir))[1:11]:
    data = pd.read_csv(datadir+"/"+f,delimiter='\t')
    print(f)
    dev_x = data['Wavelength']
    title =str(f)[:-4]
    dev_y = data[title]
    plt.plot(dev_x, dev_y)
    plt.ylabel('Reflectance')
    plt.xlabel('Wavelength')
    plt.title('f')
plt.show()
```
